## 1. Pyplot API vs Object-Oriented (OO) API
* Pyplot API â  Quickly
* OO API â Advanced

#### Pyplot API

```python
x = np.linspace(0, 10, 1000) #
plt.plot(x, np.sin(x), color="blue", linestyle="dashed", label="sin(x)"); # truyá»n vÃ o x,y, mÃ u, kiá»u ÄÆ°á»ng, nhÃ£n
plt.plot(x, np.cos(x), color="red", label="cos(x)")

plt.title("Sin(X) and Cos(X)") # tÃªn Äá» thá»
plt.xlabel("Bien X"); # tÃªn trá»¥c x: ngang
plt.ylabel("F(x)"); # tÃªn trá»¥c y: dá»c

#plt.xlim([0,4]) :giá»i háº¡n chiá»u ngang cá»§a Äá» thá»
#plt.ylim([-0.7, 0.6]) : giá»i háº¡n chiá»u dÃ i cá»§a Äá» thá»
#plt.axis([0, 4, -0.7, 0.6]) :xmin xmax ymin ymax: giá»ng 2 cÃ¡i á» trÃªn
#plt.axis("tight") :giá»i háº¡n kiá»u tight: fit vá»i Äá» thá», equal: x báº±ng y
plt.axis("equal")

# k cáº§n gá»i plt.show() chá» cáº§n gá»i ghi dÃ¹ng trÃªn ter hoáº·c pyfile
plt.legend();
```

#### Object-Oriented API

```python
%matplotlib inline
#1. Prepare data
x = [1,2,3,4]
y = [11,22,33,44]

#2. Setup plot
# fig: hÃ¬nh tá»ng, ax: tá»«ng hÃ¬nh nhá»
fig, ax = plt.subplots(figsize=(5,5)) #Figure size = Width & Height of the plot

#3. Plot the data
ax.plot(x, y);

#4. Customize the plot: gáº¯n tiÃªu Äá», x y
ax.set(title="A simple plot",
       xlabel="x-axis",
       ylabel="y-axis");
```

### 3. Bar
* Vertical: cá»t Äá»©ng
* Horizontal: cá»t ngang

```python
#Prepare Data from a python dictionary
soft_drink_prices = {"Coke": 10,
                     "Pessi": 12,
                     "Sprite": 8}

fig, ax = plt.subplots()
# biá»u Äá»: ÄÆ°a vÃ o x lÃ  key cá»§a soft_ , x lÃ  values
ax.bar(soft_drink_prices.keys(), soft_drink_prices.values(), color="pink");
ax.set(title="Bach Hoa Xanh's Soft Drink Prices", ylabel="Price ($)");
```

```python
fig, ax = plt.subplots()
# bar h: horiza.. Láº­t ngang
# hÃ m nÃ y nháº­n list chá»© k nháº­n dict nÃªn chuyá»n sang list
ax.barh(list(soft_drink_prices.keys()), list(soft_drink_prices.values()));
```

### 4. Histogram: biá»u Äá» táº§n suáº¥t dáº¡ng cá»t

### Subplots : nhiá»u Äá» thá» ( lá»£i tháº¿ cÃ¡ch OO)
* Multiple plots on one figure: https://matplotlib.org/3.1.1/gallery/recipes/create_subplots.html

```python
# Option 1:
# cho 4 Äá» thá» vá»i 2 hÃ ng 2 cá»t
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,
                                             ncols=2,
                                             figsize=(10, 5))
#Plot data to each axis
x=np.linspace(0,10,100)
ax1.plot(x, x/2);
ax1.plot(x, x*2);

ax2.scatter(np.random.random(10),np.random.random(10));
ax3.bar(soft_drink_prices.keys(), soft_drink_prices.values());
ax4.hist(student_height, bins=30);
```

```python
# Option 2:
fig, ax = plt.subplots(nrows=2,
                       ncols=2,
                       figsize=(10,5))
#Plot data to each axis
# thay vÃ¬ ax1 vÃ  gÃ¡n á» trÃªn: ta dÃ¹ng ax[0,0]: hÃ ng cá»t
ax[0,0].plot(x, x/2);
ax[0,0].plot(x, x*2);

ax[0,1].scatter(np.random.random(10),np.random.random(10));
ax[1,0].bar(soft_drink_prices.keys(), soft_drink_prices.values());
ax[1,1].hist(student_height, bins=30);
```

### Project: Data Visualization - California Cities

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
cities = pd.read_csv('/content/California Cities.csv')
cities.head()

#cities["area_total_km2"].sort_values(ascending=False) # giáº£m dáº§n
```

```python
# Extract Latd = VÄ© Äá» and Longd = Kinh Äá»

lat, lon = cities['latd'], cities['longd'] # láº¥y column latd vÃ  longd gÃ¡n vÃ o lat lon
population, area = cities['population_total'], cities['area_total_km2'] # láº¥y dÃ¢n sá», diá»n tÃ­ch

plt.style.use('seaborn') # style Äá» thá»
plt.figure(figsize=(8,6)) # size 8x6
# Plot using Pyplot API: plot kinh Äá» vÄá» lÃªn biá»u Äá» phÃ¢n tÃ¡n, c lÃ  colors nhÆ°ng mÃ  sá» nÃ³ to quÃ¡ nÃªn log10 cho nÃ³ nhá» láº¡i vá»«a biá»u Äá»
# cmap kiá»u mÃ u khÃ¡c, s: size theo area
plt.scatter(lon, lat,
            c=np.log10(population), cmap='viridis',
            s=area, linewidths=0, alpha=0.5)

plt.axis('equal') # kÄá» vÄá» kÃ©o báº±ng nhau
plt.xlabel('longlatitude')
plt.ylabel('latitude');
plt.colorbar(label='log$_{10}$(population)') # thanh mÃ u thá» hiá»n táº§n suáº¥t cá»§a mÃ u. VÃ¬ dÃ¹ng log10 nÃªn ÄÆ°a vÃ o log10
plt.clim(3,7) # giá»i háº¡n mÃ u mÃ u tá»« 3 Äáº¿n 7


#Create a legend for cities'sizes
# chia ra sau khi xem ÄÆ°á»£c sort á» trÃªn
area_range = [50, 100, 300, 500]

# k cáº§n váº½ tá»a Äá» nÃªn cho 2 list rá»ng xy
for area in area_range:
    plt.scatter([], [],  s=area, c='k', alpha=0.4,
                label=str(area) + ' km$^2$')
# khi ÄÃ£ dÃ¹ng label á» trÃªn thÃ¬ pháº£i dÃ¹ng legend nÃ³ má»i hiá»n lÃªn
# spacing= khoáº£ng cÃ¡ch cÃ¡ch label =1
plt.legend(labelspacing=1, title='City Area')


plt.title('California Cities: Population and Area Distribution');
```
