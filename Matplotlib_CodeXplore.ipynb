## 1. Pyplot API vs Object-Oriented (OO) API
* Pyplot API →  Quickly
* OO API → Advanced

#### Pyplot API

```python
x = np.linspace(0, 10, 1000) #
plt.plot(x, np.sin(x), color="blue", linestyle="dashed", label="sin(x)"); # truyền vào x,y, màu, kiểu đường, nhãn
plt.plot(x, np.cos(x), color="red", label="cos(x)")

plt.title("Sin(X) and Cos(X)") # tên đồ thị
plt.xlabel("Bien X"); # tên trục x: ngang
plt.ylabel("F(x)"); # tên trục y: dọc

#plt.xlim([0,4]) :giới hạn chiều ngang của đồ thị
#plt.ylim([-0.7, 0.6]) : giới hạn chiều dài của đồ thị
#plt.axis([0, 4, -0.7, 0.6]) :xmin xmax ymin ymax: giống 2 cái ở trên
#plt.axis("tight") :giới hạn kiểu tight: fit với đồ thị, equal: x bằng y
plt.axis("equal")

# k cần gọi plt.show() chỉ cần gọi ghi dùng trên ter hoặc pyfile
plt.legend();
```

#### Object-Oriented API

```python
%matplotlib inline
#1. Prepare data
x = [1,2,3,4]
y = [11,22,33,44]

#2. Setup plot
# fig: hình tổng, ax: từng hình nhỏ
fig, ax = plt.subplots(figsize=(5,5)) #Figure size = Width & Height of the plot

#3. Plot the data
ax.plot(x, y);

#4. Customize the plot: gắn tiêu đề, x y
ax.set(title="A simple plot",
       xlabel="x-axis",
       ylabel="y-axis");
```

```python
# cách Pyplot API
plt.scatter(x, np.exp(x)); #y=e^(x)
```

```python
# cách OO API

#Prepare New Data
rng = np.random.RandomState(0) # ra cùng kq với codexplore

x = rng.randn(100)
y = rng.randn(100)
colors = rng.rand(100) # random màu
sizes = 1000*rng.rand(100) # random size

fig, ax = plt.subplots()
img1 = ax.scatter(x,y, s=sizes, c=colors, cmap='viridis', alpha=0.3); # gán đồ thị với x,y,
# gán size là random ở trên để mỗi chấm kích thước khác nhau, màu cũng random
# cmap kiểu màu

fig.colorbar(img1); # thanh colors bar thể hiện tần số của màu
```

### 3. Bar
* Vertical: cột đứng
* Horizontal: cột ngang

```python
#Prepare Data from a python dictionary
soft_drink_prices = {"Coke": 10,
                     "Pessi": 12,
                     "Sprite": 8}

fig, ax = plt.subplots()
# biểu đồ: đưa vào x là key của soft_ , x là values
ax.bar(soft_drink_prices.keys(), soft_drink_prices.values(), color="pink");
ax.set(title="Bach Hoa Xanh's Soft Drink Prices", ylabel="Price ($)");
```

```python
fig, ax = plt.subplots()
# bar h: horiza.. Lật ngang
# hàm này nhận list chứ k nhận dict nên chuyển sang list
ax.barh(list(soft_drink_prices.keys()), list(soft_drink_prices.values()));
```

### 4. Histogram: biểu đồ tần suất dạng cột

```python
#Prepare data
np.random.seed(42) #random cùng dliệu
student_height = np.random.normal(170, 10, 250) # mean:tbinh, std: lệch chuẩn, 250 gtrị
```

```python
student_height[:10], len(student_height)
```

```python
#pyplot API
plt.hist(student_height, bins=30); # bins: chia nhỏ khoảng x ra để dễ biết từng đoạn
```

### Subplots : nhiều đồ thị ( lợi thế cách OO)
* Multiple plots on one figure: https://matplotlib.org/3.1.1/gallery/recipes/create_subplots.html

```python
# Option 1:
# cho 4 đồ thị với 2 hàng 2 cột
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,
                                             ncols=2,
                                             figsize=(10, 5))
#Plot data to each axis
x=np.linspace(0,10,100)
ax1.plot(x, x/2);
ax1.plot(x, x*2);

ax2.scatter(np.random.random(10),np.random.random(10));
ax3.bar(soft_drink_prices.keys(), soft_drink_prices.values());
ax4.hist(student_height, bins=30);
```

```python
# Option 2:
fig, ax = plt.subplots(nrows=2,
                       ncols=2,
                       figsize=(10,5))
#Plot data to each axis
# thay vì ax1 và gán ở trên: ta dùng ax[0,0]: hàng cột
ax[0,0].plot(x, x/2);
ax[0,0].plot(x, x*2);

ax[0,1].scatter(np.random.random(10),np.random.random(10));
ax[1,0].bar(soft_drink_prices.keys(), soft_drink_prices.values());
ax[1,1].hist(student_height, bins=30);
```

### Project: Data Visualization - California Cities

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
cities = pd.read_csv('/content/California Cities.csv')
cities.head()

#cities["area_total_km2"].sort_values(ascending=False) # giảm dần
```

```python
# Extract Latd = Vĩ Độ and Longd = Kinh Độ

lat, lon = cities['latd'], cities['longd'] # lấy column latd và longd gán vào lat lon
population, area = cities['population_total'], cities['area_total_km2'] # lấy dân số, diện tích

plt.style.use('seaborn') # style đồ thị
plt.figure(figsize=(8,6)) # size 8x6
# Plot using Pyplot API: plot kinh độ vđộ lên biểu đồ phân tán, c là colors nhưng mà số nó to quá nên log10 cho nó nhỏ lại vừa biểu đồ
# cmap kiểu màu khác, s: size theo area
plt.scatter(lon, lat,
            c=np.log10(population), cmap='viridis',
            s=area, linewidths=0, alpha=0.5)

plt.axis('equal') # kđộ vđộ kéo bằng nhau
plt.xlabel('longlatitude')
plt.ylabel('latitude');
plt.colorbar(label='log$_{10}$(population)') # thanh màu thể hiện tần suất của màu. Vì dùng log10 nên đưa vào log10
plt.clim(3,7) # giới hạn màu màu từ 3 đến 7


#Create a legend for cities'sizes
# chia ra sau khi xem được sort ở trên
area_range = [50, 100, 300, 500]

# k cần vẽ tọa độ nên cho 2 list rỗng xy
for area in area_range:
    plt.scatter([], [],  s=area, c='k', alpha=0.4,
                label=str(area) + ' km$^2$')
# khi đã dùng label ở trên thì phải dùng legend nó mới hiện lên
# spacing= khoảng cách cách label =1
plt.legend(labelspacing=1, title='City Area')


plt.title('California Cities: Population and Area Distribution');
```
