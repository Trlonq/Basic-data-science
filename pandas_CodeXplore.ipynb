Manipulating Data (thao tác dữ liệu)

Ví dụ giá tiền có $ thì dữ liệu nó là kiểu Object nên khó xử lý

```python
df.item_price.dtype
```

hàm Apply()

thay thế $ bằng khoảng trắng

```python
df.item_price.apply(lambda x: x.replace('$', ''))
```

nhưng tách rồi vẫn type O

dùng float() để chuyển sang số thực

```python
df.item_price.apply(lambda x: float(x.replace('$', '')))
```

test

Có thể thấy item_price vẫn chưa lưu vì ta chỉ thay đổi trên df. Muốn nó lưu thì gán nó vào df.item_price

```python
df.item_price = df.item_price.apply(lambda x: float(x.replace('$', '')))
```
